```python
## 라이브러리 불러오기

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("./data.csv")

df.head(5)
```

```python
df.drop(['Physicians','Temperature','School Closure Date','Income' , 'GDP', 'Pollution', 'Med-Large Airports','Sex Ratio','Unemployment', 'Urban' , 'Flu Deaths' , 'Age 0-25', 'Age 26-54'],axis=1,inplace=True)
```

```python
type(df.info())
```

```python
df['Infected per Population'] = df['Infected'] / df['Population']
```

```python
df['Deaths per Infected'] = df['Deaths'] / df['Infected']
```

```python
df['Tested per Population'] = df['Tested'] / df['Population']
```

```python
df.isnull().sum()
```

```python
df_des = df.describe()

df_des.drop(['Gini','Age 55+','Health Spending', 'Hospitals', 'Infected per Population', 'Deaths per Infected', 'Tested per Population'], axis = 1, inplace = True)
```
